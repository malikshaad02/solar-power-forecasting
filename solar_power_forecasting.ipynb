```python
import pandas as pd


# Create sample solar dataset
data = {
'sunlight_hours': [5, 6, 7, 8, 9, 10],
'temperature': [25, 27, 30, 32, 35, 37],
'power_generated': [20, 25, 35, 45, 55, 65]
}


df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt


plt.plot(df['sunlight_hours'], df['power_generated'])
plt.xlabel('Sunlight Hours')
plt.ylabel('Power Generated (kW)')
plt.title('Solar Power Generation')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression


X = df[['sunlight_hours', 'temperature']]
y = df['power_generated']


model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[8, 33]])
prediction
```
